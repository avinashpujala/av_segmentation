This demo illustrates decomposition of a loop-based electronic dance music track, following [1].


#### The notebook proceeds in the following steps:
<br>1. Load audio files for the complete, downmixed track, as well as for the individual loops that the track contains.
<br>2. Compute STFTs for all audio data.
<br>3. Each loop becomes a fixed template ("page") in the tensor W.
<br>The track spectrogram is the target to approximate, V.
<br>We wish to learn the activation matrix H, which answers the question
<br>"Where was each loop activated throughout % the track?"
<br>4. Visualize results.

### Initialization

```python
import os
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import IPython.display as ipd
from copy import deepcopy

from NMFToolbox.forwardSTFT import forwardSTFT
from NMFToolbox.logFreqLogMag import logFreqLogMag
from NMFToolbox.initActivations import initActivations
from NMFToolbox.NMFD import NMFD
from NMFToolbox.NMFdiag import NMFdiag
from NMFToolbox.visualizeComponentsNMF import visualizeComponentsNMF
from NMFToolbox.alphaWienerFilter import alphaWienerFilter
from NMFToolbox.utils import EPS, make_monaural, pcmInt16ToFloat32Numpy

inpPath = '../data/';
outPath = 'output/';

filename = '120_jd.wav';

filenameBass = '120_jd_bass_1.wav';
filenameMelody = '120_jd_melody_1.wav';
filenameDrums = '120_jd_drums_1.wav';
```

### 1. Load the audio signal

```python
fs, xTr = wav.read(os.path.join(inpPath, filename))
fBass, xBass = wav.read(os.path.join(inpPath, filenameBass))
fMelody, xMelody = wav.read(os.path.join(inpPath, filenameMelody))
fDrums, xDrums = wav.read(os.path.join(inpPath, filenameDrums))

# make monaural if necessary
xTr = make_monaural(xTr)
xBass = make_monaural(xBass)
xMelody = make_monaural(xMelody)
xDrums = make_monaural(xDrums)

# int16 -> float32 conversion
xTr = pcmInt16ToFloat32Numpy(xTr)
xBass = pcmInt16ToFloat32Numpy(xBass)
xMelody = pcmInt16ToFloat32Numpy(xMelody)
xDrums = pcmInt16ToFloat32Numpy(xDrums)
```

### 2. Compute STFT

```python
# spectral parameters
paramSTFT = dict()
paramSTFT['blockSize'] = 4096;
paramSTFT['hopSize'] = 2048;
paramSTFT['winFunc'] = np.hanning(paramSTFT['blockSize'])
paramSTFT['reconstMirror'] = True
paramSTFT['appendFrame'] = True
paramSTFT['numSamples'] = len(xTr)

# STFT computation
XTr, ATr, PTr = forwardSTFT(xTr, paramSTFT);

# get dimensions and time and freq resolutions
numBinsTr, numFramesTr = XTr.shape
deltaT = paramSTFT['hopSize'] / fs
deltaF = fs / paramSTFT['blockSize']

# get logarithmically-spaced frequency axis version for visualization
logFreqLogMagA, logFreqAxis = logFreqLogMag(ATr, deltaF)
numLogBins = len(logFreqAxis)

# repeat for bass loop ---------------------------------------------------
paramSTFT['numSamples'] = len(xBass);

# STFT computation
XBass, ABass, PBass = forwardSTFT(xBass, paramSTFT)

# get dimensions and time and freq resolutions
numBinsBass, numFramesBass = XBass.shape

# get logarithmically-spaced frequency axis version for visualization purposes
logFreqLogMagABass, logFreqAxisBass = logFreqLogMag(ABass, deltaF)
numLogBinsBass = len(logFreqAxisBass)

# repeat for Melody loop ---------------------------------------------------
paramSTFT['numSamples'] = len(xMelody)

# STFT computation
XMelody, AMelody, PMelody = forwardSTFT(xMelody, paramSTFT)

# get dimensions and time and freq resolutions
numBinsMelody, numFramesMelody = XMelody.shape

# get logarithmically-spaced frequency axis version for visualization purposes
logFreqLogMagAMelody, logFreqAxisMelody = logFreqLogMag(AMelody, deltaF)
numLogBinsMelody = len(logFreqAxisMelody)

# repeat for Drums loop ---------------------------------------------------
paramSTFT['numSamples'] = len(xDrums)

# STFT computation
XDrums, ADrums, PDrums = forwardSTFT(xDrums, paramSTFT)

# get dimensions and time and freq resolutions
numBinsDrums, numFramesDrums = XDrums.shape

# get logarithmically-spaced frequency axis version for visualization purposes
logFreqLogMagADrums, logFreqAxisDrums = logFreqLogMag(ADrums, deltaF)
numLogBinsDrums = len(logFreqAxisDrums)
```

### 3. Apply NMF variants to STFT magnitude

```python
# set common parameters
numComp = 3
numIter = 30
numTemplateFrames = deepcopy(numFramesBass)

# generate initial activations
paramActivations = dict()
paramActivations['numComp'] = numComp
paramActivations['numFrames'] = numFramesTr
initH = initActivations(paramActivations, 'uniform')
```

```python
# NMFD parameters
paramNMFD = dict()
paramNMFD['numComp'] = numComp
paramNMFD['numFrames'] = numFramesTr
paramNMFD['numIter'] = numIter
paramNMFD['numTemplateFrames'] = numTemplateFrames

paramNMFD['initH'] = initH
paramNMFD['numBins'] = numBinsTr

# NMFD core method
nmfdW, nmfdH, nmfdV, divKL, _ = NMFD(ATr, paramNMFD)

# alpha-Wiener filtering
nmfdA, _ = alphaWienerFilter(ATr, nmfdV, 1)

# make log freq version of component spectrograms
logFreqLogMagNmfdA = logFreqLogMag(nmfdA, deltaF)
```

```python
# visualize
plt.figure();
plt.subplot(3,1,1)
plt.plot(nmfdH[0, :])
plt.subplot(3,1,2)
plt.plot(nmfdH[1, :])
plt.subplot(3,1,3);
plt.plot(nmfdH[2, :])
```

### 3. Apply continuity NMF variants to mosaicing pair

```python
numSourceFrames = numComp * numTemplateFrames
numTargetFrames = numFramesTr

# initialize activations randomly
# generate initial activations
paramActivations['numComp'] = numSourceFrames
paramActivations['numFrames'] = numFramesTr
H0 = initActivations(paramActivations, 'random')

As = np.concatenate([ADrums, AMelody, ABass], axis=1)
Xs = np.concatenate([XDrums, XMelody, XBass], axis=1)
```

```python
# init templates by source frames
W0 = As * 1 / (EPS + As.sum(axis=0))
Xs = Xs * 1 / (EPS + Xs.sum(axis=0))

# parameters taken from Jonathan Driedger's toolbox
paramNMFdiag = dict()
paramNMFdiag['fixW'] = True
paramNMFdiag['numOfIter'] = 20
paramNMFdiag['continuity'] = dict()
paramNMFdiag['continuity']['polyphony'] = 1
paramNMFdiag['continuity']['length'] = np.floor(numTemplateFrames/2).astype(np.int)
paramNMFdiag['continuity']['grid'] = 1
paramNMFdiag['continuity']['sparsen'] = [1, numTemplateFrames * 2 + 1]


# call the reference implementation as provided by Jonathan Driedger
nmfdiagW, nmfdiagH = NMFdiag(ATr, W0, H0, paramNMFdiag)
```

```python
plt.figure()
plt.imshow(nmfdiagH, origin='lower', aspect='auto', cmap='gray_r')
```

```python
# create mosaic, replacing the magnitude frames by complex valued frames
contY = np.matmul(Xs, nmfdiagH)

# visualize
paramVis = dict()
paramVis['deltaF'] = deltaF
paramVis['deltaT'] = deltaT
fh1, _ = visualizeComponentsNMF(ATr, nmfdiagW, nmfdiagH, None, paramVis)
```
