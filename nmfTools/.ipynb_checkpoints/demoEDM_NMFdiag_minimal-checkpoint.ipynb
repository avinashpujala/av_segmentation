```python
import os
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import IPython.display as ipd
from copy import deepcopy

from NMFToolbox.forwardSTFT import forwardSTFT
from NMFToolbox.inverseSTFT import inverseSTFT
from NMFToolbox.logFreqLogMag import logFreqLogMag
from NMFToolbox.initActivations import initActivations
from NMFToolbox.NMFD import NMFD
from NMFToolbox.NMFdiag import NMFdiag
from NMFToolbox.visualizeComponentsNMF import visualizeComponentsNMF
from NMFToolbox.alphaWienerFilter import alphaWienerFilter
from NMFToolbox.utils import EPS, make_monaural, pcmInt16ToFloat32Numpy

inpPath = '../data/';
outPath = 'output/';

filename = '120_jd.wav';

filenameBass = '120_jd_bass_1.wav';
filenameMelody = '120_jd_melody_1.wav';
filenameDrums = '120_jd_drums_1.wav';
```

### 1. Load the Audio signal

```python
fs, xTr = wav.read(os.path.join(inpPath, filename))
fBass, xBass = wav.read(os.path.join(inpPath, filenameBass))
fMelody, xMelody = wav.read(os.path.join(inpPath, filenameMelody))
fDrums, xDrums = wav.read(os.path.join(inpPath, filenameDrums))

# make monaural if necessary
xTr = make_monaural(xTr)
xBass = make_monaural(xBass)
xMelody = make_monaural(xMelody)
xDrums = make_monaural(xDrums)

# int16 -> float32 conversion
xTr = pcmInt16ToFloat32Numpy(xTr)
xBass = pcmInt16ToFloat32Numpy(xBass)
xMelody = pcmInt16ToFloat32Numpy(xMelody)
xDrums = pcmInt16ToFloat32Numpy(xDrums)
```

### 2. Compute STFT

```python
# spectral parameters
paramSTFT = dict()
paramSTFT['blockSize'] = 4096;
paramSTFT['hopSize'] = 2048;
paramSTFT['winFunc'] = np.hanning(paramSTFT['blockSize'])
paramSTFT['reconstMirror'] = True
paramSTFT['appendFrame'] = True
paramSTFT['numSamples'] = len(xTr)

# STFT computation
XTr, ATr, PTr = forwardSTFT(xTr, paramSTFT);

# get dimensions and time and freq resolutions
numBinsTr, numFramesTr = XTr.shape
deltaT = paramSTFT['hopSize'] / fs
deltaF = fs / paramSTFT['blockSize']

# get logarithmically-spaced frequency axis version for visualization
logFreqLogMagA, logFreqAxis = logFreqLogMag(ATr, deltaF)
numLogBins = len(logFreqAxis)

# repeat for bass loop ---------------------------------------------------
paramSTFT['numSamples'] = len(xBass);

# STFT computation
XBass, ABass, PBass = forwardSTFT(xBass, paramSTFT)
numTemplateFrames = ABass.shape[1]

# repeat for Melody loop ---------------------------------------------------
paramSTFT['numSamples'] = len(xMelody)

# STFT computation
XMelody, AMelody, PMelody = forwardSTFT(xMelody, paramSTFT)

# repeat for Drums loop ---------------------------------------------------
paramSTFT['numSamples'] = len(xDrums)

# STFT computation
XDrums, ADrums, PDrums = forwardSTFT(xDrums, paramSTFT)
```

```python
As = np.concatenate([ADrums, AMelody, ABass], axis=1)
W0 = As * 1 / (EPS + As.sum(axis=0))
numSourceFrames = W0.shape[1]
numTargetFrames = numFramesTr

# initialize activations randomly
paramActivations = dict()
paramActivations['numComp'] = numSourceFrames
paramActivations['numFrames'] = numTargetFrames
H0 = initActivations(paramActivations, 'random')
```

### 3. Apply continuity NMF variants to STFT magnitude

```python
# parameters taken from Jonathan Driedger's toolbox
paramNMFdiag = dict()
paramNMFdiag['fixW'] = True
paramNMFdiag['numOfIter'] = 20
paramNMFdiag['continuity'] = dict()
paramNMFdiag['continuity']['polyphony'] = 1
paramNMFdiag['continuity']['length'] = np.floor(numTemplateFrames/2).astype(np.int)
paramNMFdiag['continuity']['grid'] = 1
paramNMFdiag['continuity']['sparsen'] = [1, numTemplateFrames * 2 + 1]

# call the reference implementation as provided by Jonathan Driedger
nmfdiagW, nmfdiagH = NMFdiag(ATr, W0, H0, paramNMFdiag)
    
nmfdiagV = nmfdiagW @ nmfdiagH
```

```python
# visualize
paramVis = dict()
paramVis['deltaF'] = deltaF
paramVis['deltaT'] = deltaT
fh1, _ = visualizeComponentsNMF(nmfdiagV, nmfdiagW, nmfdiagH, None, paramVis)
```
